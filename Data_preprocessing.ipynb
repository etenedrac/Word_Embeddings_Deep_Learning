# Data preprocessing of the corpus

```python
import json
```

```python
text_total = []
```

```python
with open('./data/recipes_raw/recipes_raw_nosource_fn.json') as json_file:  
    data = json.load(json_file)
    for d in data:
        try:
            print(data[d]['title'])
            recipe = data[d]['title']+ " " + data[d]['instructions']
            text_total.append(recipe)
        except:
            # Key is not present
            pass
```

```python
import pickle

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
text_processed = []
table_line = str.maketrans({"\n": "  "})
table_punct = str.maketrans({key: None for key in string.punctuation})
for index, recipe in enumerate(text_total):
    text_processed.append(text_total[index].translate(table_punct).translate(table_line).lower())
```
